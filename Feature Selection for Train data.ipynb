```python
import pandas as pd
import numpy as np
```

```python
# Feature Selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
df = pd.read_csv('x_train.csv')
df.head()
```

```python
# independent feature
X_train = df.drop('SalePrice', axis=1)

# dependent feature
y_train = df['SalePrice']
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).

# The bigger the alpha the less features that will be selected.

feature_selection = SelectFromModel(Lasso(alpha=0.005, random_state=0)) 
# set random state for test data too

feature_selection.fit(X_train, y_train)
```

```python
feature_selection.get_support()
```

```python
selected_features = X_train.columns[(feature_selection.get_support())]

print('Total Number of Features: {}'.format(len(X_train.columns)))
print('Selected Features {}'.format(len(selected_features)))
print('Number of Features with coefficients shrank to zero: {}'.format(np.sum(feature_selection.estimator_.coef_==0)))
```
